```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("similar_excel_done.csv", delimiter=';', index_col='ind')
```

```python
df.head()
```

```python
df = df.drop(df[df['keep name'].isin([np.nan, 'k'])].index)
```

```python
df.head()
```

```python
new_columns = df.columns.values
new_columns[1] = '1'
new_columns[3] = '2'
new_columns[5] = '3'
new_columns[7] = '4'
df.columns = new_columns
```

```python
df.head()
```

```python
unique_names = {}
for row in df.iterrows():
    name = df['orig_name'].loc[row[0]]
    unique_names[name] = []
    num_names = df['keep name'].loc[row[0]].split('-')
    for n in num_names:
        unique_names[name].append(df[n].loc[row[0]])
```
